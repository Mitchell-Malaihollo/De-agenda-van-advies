```python
def parse_python_list(text):
    try:
        val = ast.literal_eval(text.strip())
        return val if isinstance(val, list) else []
    except Exception:
        return []

def eersteprompt(titel, adviescollege, datum, bron):
    client = OpenAI()
    prompt = f"""
Jij bent een onderzoeksassistent gespecialiseerd in het vinden van Nederlandse overheidsdocumenten en parlementaire stukken.

Advies:
- titel: {titel}
- adviescollege: {adviescollege}
- publicatiedatum: {datum}
- bronpagina advies: {bron}

Taak:
Vind zo veel mogelijk overheidsdocumenten die dit advies expliciet of impliciet noemen of behandelen (parlementaire stukken, Kamerbrieven, Kamervragen+antwoorden, kabinetsreacties/standpunten, beleidsnota's, wetsvoorstellen+MvT/MvA, evaluaties/monitoring, inspectie- of rekenkamerrapporten, begroting/verantwoording, uitvoeringsdocumenten, Woo-stukken).

Zoekstrategie (belangrijk):
- Gebruik meerdere zoekvarianten met deze ankers:
  1) exacte titel tussen aanhalingstekens: "{titel}"
  2) titel zonder aanhalingstekens + adviescollege: {titel} {adviescollege}
  3) alleen kenmerkende woorden uit de titel + adviescollege
  4) bron-URL domein + titelwoorden
- Prioriteer resultaten van: overheid.nl, tweedekamer.nl, officielebekendmakingen.nl, rijksoverheid.nl en pdf's van ministeries/uitvoeringsorganisaties.

Output (STRICT):
- Geef uitsluitend een Python-lijst terug in de vorm ["...", "..."].
- Elke string is één URL naar een document/pagina waar het advies genoemd wordt.
- Geen extra tekst/markdown.
"""
    
    response = client.responses.create(
        model="gpt-5.1",
        input=prompt,
        tools=[{"type": "web_search"}],
        max_output_tokens=800,
    )
    return response.output_text

def tweedeprompt(checkURL, titel, adviescollege, datum, bron):
    client = OpenAI()
    prompt = f"""
    mijn gekozen advies:
    titel advies: {titel}
    adviescollege: {adviescollege}
    Datum advies publicaties: {datum}
    webpagina advies: {bron}
    
    controleer of deze link "{checkURL}" daadwerkelijk mijn gekozen advies impliciet of expliciet benoemd.
    als de link mijn gekozen advies impliciet of expliciet noemt alleen het woordje TRUE teruggeven.
    als de link mijn gekozen advies niet noemt of de link niet werkt alleen het woordje FALSE teruggeven.
    niks anders terug geven dan TRUE of FALSE.
    """
    response = client.responses.create(
        model="gpt-5.1",
        input=prompt,
        tools=[{"type": "web_search"}],
        max_output_tokens=800,
    )
    return response.output_text.strip()

def derdeprompt(URL):
    client = OpenAI()
    prompt = f"""
    geef een naam voor dit document: {URL}

    Belangrijk:
    - geef alleen de naam, niks anders.
    - maak de naam niet te lang.
    - voorbeelden van namen voor documenten die je eventueel kan gebruiken maar dat hoeft niet:
    - - Kabinetsreactie
    - - samenvatting
    - - nieuwsartikel
    - - evaluatie
    - - advies
    - - rapport
    - - adviesaanvraag
    """
    response = client.responses.create(
        model="gpt-5.1",
        input=prompt,
        tools=[{"type": "web_search"}],
        max_output_tokens=800,
    )
    return response.output_text.strip()
```
