#### Kwalitatief score schema

Scores:
- -1 = Samenvatting mist
- 0 = Niet correct; Uitspraken zijn niet herleidbaar tot de bron. Aanwezigheid duidelijke inconsistenties, toevoegingen of verdraaiingen. 
- 0.5 = Gedeeltelijk correct; Uitspraken bevatten lichte interpretatie. Geen feitelijke fouten. Niet expliciete maar wel logische inconsistenties, toevoegingen of verdraaiingen.
- 1 = Volledig correct; Uitspraken zijn direct herleidbaar. Geen inconsistenties, toevoegingen of verdraaiingen.
